```python
driver = webdriver.Chrome()  

driver.get(url)

driver.implicitly_wait(20)
html = driver.page_source

time_regex = r"\d{1,2}:\d{2}"
soup = BeautifulSoup(html, "html.parser")
events = []

title_list = []
date_list = []
start_time_list = []
end_time_list = []

for event_tag in soup.find_all("div", class_="jsx-401363670 eapp-events-calendar-name-component eapp-events-calendar-masonry-item-name"):
    title = event_tag.text.strip()
    title_list.append(title)

time_tags = soup.find_all("script", type="application/ld+json")
for tag in time_tags:
    if "endDate" in json.loads(tag.text.strip()):
        try:
            event_date = json.loads(tag.text.strip())["endDate"]
            date_list.append(event_date)
        except:
            pass


startime_tags = soup.find_all("div", class_="jsx-1364132950 eapp-events-calendar-time-text")
for tag in startime_tags:
    times = tag.text.strip()
    try:
        # Extract start time
        start_time = times.split(" - ")[0]
        if "PM" in start_time:
            start_time = str(int(start_time.split(":")[0]) + 12) + f":{start_time.split(':')[1]}"
            start_time = start_time.replace("PM", "").strip()
            start_time = re.findall(time_regex, start_time)[0]
            start_time_list.append(start_time)

        elif "AM" in start_time:
            start_time = start_time.replace("AM", "").strip()
            start_time = re.findall(time_regex, start_time)[0]
            start_time_list.append(start_time)
        else:
            start_time = re.findall(time_regex, start_time)[0]
            start_time_list.append(start_time)

    except:
        start_time = "None"
        start_time_list.append(start_time)

    try:
        end_time = times.split(" - ")[1]
        if "PM" in end_time:
            end_time = str(int(end_time.split(":")[0]) + 12) + f":{end_time.split(':')[1]}"
            end_time = end_time.replace("PM", "").strip()
            end_time = re.findall(time_regex, end_time)[0]
            end_time_list.append(end_time)
            
        elif "AM" in end_time:
            end_time = end_time.replace("AM", "").strip()
            end_time = re.findall(time_regex, end_time)[0]
            end_time_list.append(end_time)

        else:
            end_time = re.findall(time_regex, end_time)[0]
            end_time_list.append(end_time)
            
    except:
        end_time = "None"
        end_time_list.append(end_time)

for title, event_date, start_time, end_time in zip(title_list, date_list, start_time_list, end_time_list):
    events.append({
        "title": title,
        "date": event_date,
        "start_time": start_time,
        "end_time": end_time,
        "href": "https://www.accj.or.jp/accj-events"
    })

for event in events[:]:  
    if datetime.datetime.strptime(event["date"], "%Y-%m-%d") < datetime.datetime.now():
        events.remove(event)

for event in events:
    print(f"Title: {event['title']}")
    print(f"Date: {event['date']}")
    print(f"Start Time: {event['start_time']}")
    print(f"End Time: {event['end_time']}")
    print(f"Href: {event['href']}")

driver.quit()
```

```python
import requests
from requests.auth import HTTPBasicAuth
from datetime import datetime
import os

class EventCrawlerPipeline:
    def __init__(self):
        self.wp_url = 'https://www.diplomacyandbusiness.tokyo/wp-json/tribe/events/v1/events'
        self.headers = {
            'Content-Type': 'application/json',
            'Accept': '*/*',
            'Connection': 'keep-alive',
            'User-Agent': 'PostmanRuntime/7.42.0'
        }
        self.wp_user = os.getenv('WP_USER', 'Crawler')
        self.wp_password = os.getenv('WP_PASSWORD', 'sA1X oqEA yKRS 3L4e b4GT 3OtU')
        self.auth = HTTPBasicAuth(self.wp_user, self.wp_password)
            
    def open_spider(self, spider):
        spider.logger.info("Pipeline abierto y listo para enviar eventos a la API.")

    def event_exists_in_wp(self, title, start_date):
        """Check if an event with the given title and start date already exists."""
        page = 1
        formatted_start_date = start_date.split("T")[0]  # Format date part only for comparison

        while True:
            try:
                response = requests.get(f"{self.wp_url}?page={page}", headers=self.headers, auth=self.auth)
                if response.status_code == 200:
                    events = response.json().get('events', [])
                    if not events:
                        break  # End of pagination

                    for event in events:
                        event_start_date = event['start_date'].split(" ")[0]  # Align format for comparison
                        if event['title'].strip().lower() == title.lower() and event_start_date == formatted_start_date:
                            print(f"Duplicate found: '{title}' on {start_date}")
                            return True  # Event exists
                elif response.status_code == 404:
                    break  # No more event pages
                else:
                    print(f"Failed to fetch events: {response.status_code} - {response.text}")
                    break
                page += 1
            except Exception as e:
                print(f"Error fetching events: {e}")
                break
        return False
    
    def format_datetime_iso(self, date_string, time_string="00:00:00"):
        try:
            if "@" in date_string:
                current_year = datetime.now().year
                date_time_str = f"{date_string} {current_year}"
                parsed_datetime = datetime.strptime(date_time_str, '%B %d @ %I:%M %p %Y')
                return parsed_datetime.strftime('%Y-%m-%dT%H:%M:%S')  # ISO format

            date = datetime.strptime(date_string.strip(), '%Y-%m-%d')
            time = datetime.strptime(time_string.strip(), '%H:%M:%S').time()
            combined = datetime.combine(date, time)
            return combined.strftime('%Y-%m-%dT%H:%M:%S')
        except (ValueError, AttributeError) as e:
            print(f"Error formatting date/time: {e}")
            return None
    
    def process_item(self, item, spider):
        title = item['title'].strip()
        start_date = self.format_datetime_iso(item['start_date'], item['start_time'])
        end_date = self.format_datetime_iso(item.get('end_date', item['start_date']), item.get('end_time', item['start_time']))

        if self.event_exists_in_wp(title, start_date):
            spider.logger.info(f"Duplicate event '{title}' with start date '{start_date}' detected. Skipping upload.")
            return item

        event_data = {
            'title': title,
            'description': item['description'],
            'start_date': start_date,
            'end_date': end_date,
            'all_day': False,
            'timezone': 'Asia/Tokyo',
            'cost': item.get('cost', ''),
            'website': item.get('registration_url'),
            'venue': {
                'venue': item.get('location', ''),
                'address': item.get('address', ''),
                'city': item.get('city', ''),
                'country': item.get('country', ''),
                'state': item.get('state', ''),
                'zip': item.get('zip', ''),
                'phone': item.get('phone', ''),
                'website': item.get('venue_website', ''),
            },
            'organizer': [{
                'organizer': item.get('organizer', ''),
                'phone': item.get('organizer_phone', ''),
                'website': item.get('organizer_website', ''),
                'email': item.get('organizer_email', ''),
            }],
            'image': item.get('image_url'),
            'show_map': True,
            'show_map_link': True
        }

        try:
            create_response = requests.post(
                self.wp_url,
                json=event_data,
                headers=self.headers,
                auth=self.auth
            )
            if create_response.status_code != 201:
                spider.logger.error(f"Failed to create event: {create_response.text}")
            else:
                spider.logger.info(f"Event created successfully: {create_response.json().get('id')}")
        except Exception as e:
            spider.logger.error(f"Error en la solicitud de creaciÃ³n de evento: {e}")

        return item
    
    def close_spider(self, spider):
        spider.logger.info("Pipeline closed. No more events will be sent to the API.")
```
